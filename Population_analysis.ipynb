```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('C:/Users/adith/Documents/SCT/population.csv',encoding='utf-8')
```

```python
df.drop(['Country Code'], axis=1, inplace=True)
```

```python
df.head()
```

```python

Country = 'India'
countryData = df[df['Country Name'] == Country]


years = countryData['Year'].astype(int)  
population = countryData['Value'].astype(int) 

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(years, population, marker='o')
plt.title(f"Population of {Country} Over Time")
plt.xlabel("Year")
plt.ylabel("Population")
plt.grid(True)
plt.show()
```

```python
plt.hist(population,bins=20,color='green',edgecolor='black')
plt.show()
```
